Distribution of Smartphone Models

```python
brand_counts = df['brand_name'].value_counts()

total_count = brand_counts.sum()

small_categories = brand_counts[brand_counts / total_count < 0.03]

brand_counts['Other'] = brand_counts[small_categories.index].sum()
brand_counts.drop(index=small_categories.index, inplace=True)

plt.figure(figsize=(10, 6))
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Smartphone Models')
plt.axis('equal') 
plt.show()
```

showing best 5 phones based on rating

```python
import matplotlib.pyplot as plt


top_5_models = df.nlargest(5, 'rating')


model_names = top_5_models['model']  
ratings = top_5_models['rating']

colors = ['blue', 'green', 'red', 'cyan', 'magenta']

plt.figure(figsize=(10, 6))
plt.barh(model_names, ratings, color=colors)
plt.xlabel('Rating')
plt.ylabel('Model')
plt.title('Top 5 Smartphone Models by Rating')
plt.gca().invert_yaxis() 
plt.show()
```

Ratings regard to the PRICE

```python
ratings = df['rating']
prices = df['price']


plt.figure(figsize=(10, 6))
plt.scatter(ratings, prices, color='blue', alpha=0.6)
plt.xlabel('Rating')
plt.ylabel('Price')
plt.title('Scatter Plot of Smartphone Ratings')
plt.grid(True)
plt.show()
```

plotting PROCESSOR_SPEED rather PRICE

```python
processor_speed = df['processor_speed']
prices = df['price']


plt.figure(figsize=(10, 6))
plt.scatter(processor_speed, prices, color='blue', alpha=0.6)
plt.xlabel('Processor_speed')
plt.ylabel('Price')
plt.title('Scatter Plot of Smartphone Ratings')
plt.grid(True)
plt.show()
```

plotting proceccor speed ragard to price

```python

df['processor_speed_bin'] = pd.cut(df['processor_speed'], bins=5)

avg_prices = df.groupby('processor_speed_bin')['price'].mean().reset_index()

processor_speed_bins = avg_prices['processor_speed_bin'].astype(str)  
prices = avg_prices['price']

plt.figure(figsize=(10, 6))
plt.bar(processor_speed_bins, prices, color='blue')
plt.xlabel('Processor Speed Intervals')
plt.ylabel('Average Price')
plt.title('Average Price by Processor Speed ')
plt.grid(True)
plt.xticks(rotation=45) 
plt.show()
```
